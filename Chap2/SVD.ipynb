```julia
using PlotlyJS
using LinearAlgebra
output = true
```

```julia
A = [1 2; -1 3]
U, S, W = svd(A)
t = LinRange(0,2,129)
r = [cos.(t)'; sin.(t)']
Ar = A*r
Br = diagm(0 => S) * W' * r

t1 = scatter(x=r[1,:],y=r[2,:],name="Unit circle")
t2 = scatter(x=Ar[1,:],y=Ar[2,:],name="Map of unit circle")
t3 = scatter(x=[U[1,1],0,U[1,2]],y=[U[2,1],0,U[2,2]],name="Left singular vectors")
t4 = scatter(x=[W[1,1],0,W[1,2]],y=[W[2,1],0,W[2,2]],name="Right singular vectors")
t5 = scatter(x=Br[1,:],y=Br[2,:],name="Ellipse")
la = Layout(yaxis_scaleanchor="x", width=500, height=300)
p = plot([t1, t4, t5, t3, t2],la)
if output
    PlotlyJS.savefig(p,"fig_la10.pdf")
end
display(p)
```

```julia
e = 0.25
V = [1 1; -e e]
D = diagm(0 => [1,-1])
A = V * D / V
U, S, W = svd(A)
t = LinRange(0,2,129)
r = [cos.(t)'; sin.(t)']
Ar = A*r

t1 = scatter(x=r[1,:],y=r[2,:],name="Unit circle")
t2 = scatter(x=Ar[1,:],y=Ar[2,:],name="Map of unit circle")
t3 = scatter(x=[U[1,1],0,U[1,2]],y=[U[2,1],0,U[2,2]],name="Left singular vectors")
t4 = scatter(x=[W[1,1],0,W[1,2]],y=[W[2,1],0,W[2,2]],name="Right singular vectors")
V *= 2
t5 = scatter(x=[V[1,1],0,V[1,2]],y=[V[2,1],0,V[2,2]],name="Eigenvectors")
la = Layout(yaxis_scaleanchor="x", width=700, height=300)
p = plot([t1, t2, t3, t4, t5],la)
if output
    PlotlyJS.savefig(p,"fig_la9.pdf")
end
display(p)
```
